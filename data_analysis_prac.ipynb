```python
def popularity_measure(x):
  if x < 220:
    return "Relatively Unknown"
  elif x < 600:
    return "Kind Of Popluar"
  elif x < 2000:
    return "popular"
  else:
    return "Super Popular"
```

```python
players["popularity"] = players["page_views"].apply(popularity_measure)
```
